```python
import matplotlib
import timeit 
import random
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import set_matplotlib_formats
```

```python
'Function list_processing return element of the list'
def list_processing(arr,index):
    return arr[index]
```

### Printed results

```python
'Creation for-loop in order to measure time, increasing size of list'
for i in range(10_000, 5_000_001, 250_000):
    
    'Creation a list, using the most time-efficient methods' 
    'according to Chapter 2.6 of the book Problem Solving with Algorithms and Data Structures Using Python'
    array = list(range(i)) 
    
    'Choose random index for further selection'
    index = random.randint(0,i-1)
    
    'Using a Timer object, timing measurements by running functions list_processing'
    time_list = timeit.Timer("list_processing(array,index)", "from __main__ import list_processing, array, index")
     
        

    'Print results, depending on list size'
    print('Selection of an element out of {} is required {:.6f} for list'
          .format(i, time_list.timeit(number=1)))
```

### Visualisation 

```python
'Initialization of y-axis'
y = []

'Index initialization as constant, in order to make the plot smoother'
index = 1

'Creation for-loop in order to measure time, increasing size of list'
for i in range(10_000, 2_000_001, 30_000):
    array = list(range(i))
    
    time_list = timeit.Timer("list_processing(array,index)", "from __main__ import list_processing, array, index")

    'Append results for further visualisation'
    y.append(time_list.timeit(number=1)*10**6)
```

```python
'Initialization of x-axis'
x = np.arange(10000, 2_000_001, 30_000)

'Description of axises, sizes, labels'
fig, ax = plt.subplots()
plt.ylim(ymin=-1, ymax=100)
plt.xlim(xmin=10000, xmax=2000001)
plt.ylabel("Running Time (microsecond)")
plt.xlabel("Size of input (number of elements)")
plt.xticks(np.arange(0,2000001, 500000))

'Using markers for better visualisation, without lines'
ax.plot(x, y, label='Time complicity for operator $list[index]$',  marker = '^', linestyle='', color='g')


legend = plt.legend(loc='upper center', fontsize='medium')

plt.show()
```
